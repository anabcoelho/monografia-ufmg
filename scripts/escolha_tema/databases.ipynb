# German Bank NO OK

https://archive.ics.uci.edu/dataset/144/statlog+german+credit+data

negativo, muitas variáveis categóricas -> e da trabalho interpretar

```python
from ucimlrepo import fetch_ucirepo

# Fetch dataset
statlog_german_credit_data = fetch_ucirepo(id=144)

# Dados
X = statlog_german_credit_data.data.features
y = statlog_german_credit_data.data.targets

# Informações das variáveis
variables = statlog_german_credit_data.variables
# Extrair nomes das variáveis de entrada
feature_names = variables[variables["role"] == "Feature"]["name"].tolist()

# Atribuir como colunas
X.columns = feature_names
df = pd.concat([X, y], axis=1)
print(df.shape)
df.head()
```

```python
import pandas as pd

def read_csv_robusto(path):
    encodings = ["utf-8", "latin1", "ISO-8859-1", "cp1252"]

    for enc in encodings:
        try:
            return pd.read_csv(path, encoding=enc)
        except UnicodeDecodeError:
            continue

    # último recurso: ignorar caracteres inválidos
    return pd.read_csv(path, encoding="utf-8", encoding_errors="ignore")
```

```python
import os
import pandas as pd
import sweetviz as sv

data_dir = extract_path
csv_files = [f for f in os.listdir(data_dir) if f.lower().endswith(".csv")]

for csv_file in csv_files:
    csv_path = os.path.join(data_dir, csv_file)
    print(f"\nLendo {csv_file}...")

    try:
        df = read_csv_robusto(csv_path)
        print("Shape:", df.shape)

        # Já rodei a análise
        #report = sv.analyze(df)

        #html_name = csv_file.replace(".csv", "_sweetviz.html")
        #html_path = os.path.join(data_dir, html_name)

        #report.show_html(html_path, open_browser=False)
        #print(f"Relatório salvo em: {html_path}")

    except Exception as e:
        print(f"❌ Erro em {csv_file}: {e}")
```
